### 1. Prime in a Power Set

```
def isprime(num):
  if(num>1):
   for i in range(2,num):
       if(num%i)==0:
           return False
   else:
           return True
  else:
      return False
```

```
isprime(31)
```

```
def findsubsets(s): 
    lst = []
    for i in range(1, len(s)+1):
      lst.append(list(itertools.combinations(s, i)))
    return lst

def printprime(s):
  lst = findsubsets(s)
  primes = []
  for i in lst:
    for j in i:
      if(isprime(sum(list(j)))):
        primes.append(j)
  return primes
```

```
arr = []
for i in range(1,19):
  arr.append(i)
arr
```

```
%%time
print(printprime(arr))
```

```
%%time
def worker1(arr): 
  return printprime(arr)


def worker2(arr): 
  return printprime(arr)

if __name__ == "__main__": 

  # creating processes 
  p1 = multiprocessing.Process(target=worker1) 
  p2 = multiprocessing.Process(target=worker2)

  # starting processes 
  p1.start() 
  p2.start()

  # wait until processes are finished 
  p1.join() 
  p2.join() 

  print(worker1(arr[:9]))
  print(worker2(arr[9:]))
  print("Done!") 
```

### 2. Word Search in English Dictionary

```
!wget https://raw.githubusercontent.com/dwyl/english-words/master/words.txt
```

```
def isPalindrome(s):
    return s == s[::-1]
```

```
f = open('words.txt', 'r') 
Lines = f.readlines() 
palin = []
for line in Lines: 
    line = line.strip()
    if(isPalindrome(line)):
      palin.append(line)
```

```
%%time
print(palin)
```

```
len(Lines)//2
```

```
%%time
def worker1(Lines): 
  palin = []
  for line in Lines: 
    line = line.strip()
    if(isPalindrome(line)):
      palin.append(line)
  return palin


def worker2(Lines): 
  palin = []
  for line in Lines: 
    line = line.strip()
    if(isPalindrome(line)):
      palin.append(line)
  return palin

if __name__ == "__main__": 

  # creating processes 
  p1 = multiprocessing.Process(target=worker1) 
  p2 = multiprocessing.Process(target=worker2)

  # starting processes 
  p1.start() 
  p2.start()

  # wait until processes are finished 
  p1.join() 
  p2.join() 

  print(worker1(Lines[:233275]))
  print(worker2(Lines[233275:]))
  print("Done!") 
```

```
%%time
def worker1(image): 
  return scipy.ndimage.convolve(image, kernel3d)

def worker2(image): 
  return scipy.ndimage.convolve(image, kernel3d)

if __name__ == "__main__": 

  # creating processes 
  p1 = multiprocessing.Process(target=worker1) 
  p2 = multiprocessing.Process(target=worker2)

  # starting processes 
  p1.start() 
  p2.start()

  # wait until processes are finished 
  p1.join() 
  p2.join() 

  print(worker1(image[:1500,:1200,:]))
  print(worker2(image[1500:,1200:,:]))
  print("Done!") 
```
